```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load dataset
data = pd.read_csv("/content/AmesHousing.csv")
print(data.info())
```

```python
# Histogram creation
plt.figure(figsize = (10,6))
plt.hist(data['SalePrice'], edgecolor = "k")
plt.xlabel("Saleprice")
plt.ylabel("Frequency")
plt.title("Distribution of Sale prices")
plt.show

# Box plot compare Saleprice distributions across Neighborhoods
plt.figure(figsize=(25,14))
sns.boxplot(x = 'Neighborhood', y = 'SalePrice', data = data,)
plt.xlabel('Neighborhood')
plt.ylabel('Saleprice')
plt.title('Sale Price Distribution Across Neighborhoods')
plt.grid()
plt.show()

# Scatter plot to show the relationship btw property size and price
plt.scatter(data['Gr Liv Area'], data['SalePrice'], alpha = 0.5, edgecolor = 'w', linewidth = 0.5)
plt.xlabel("Area")
plt.ylabel("Saleprice")
plt.title("Relationship Between Property Size and Sale Price")
plt.grid()
plt.show()
```
